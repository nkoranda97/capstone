## Data Processing

This takes the pylidc objects and builds our dataset. Details can be found in the corresponding python file, but for a given scan, it goes through the following process:

1. Creates train/val splits on patient level.
2. Creates a union of the annotations to create the max area bounding box and converts to yolo format.
3. Identifies the slice corresponding to the middle z coordinate of a node. 
4. Chooses negative slices far away from nodes.
5. Creates 3 channel representation of positives (middle k slice, &plusmn; 1 k) and negative slice. Converts to JPEG.
6. Creates a YAML file with metadata for dataset organization used by YOLO, and a CSV file for metadata on individual image splits.

```python
scans = pl.query(pl.Scan).all()

build_dataset(
    scans=scans,
    root_dir="../data/yolo",
)
```
