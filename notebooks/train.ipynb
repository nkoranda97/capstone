```python
from ultralytics import YOLO

import torch

print("cuda available:", torch.cuda.is_available())
if torch.cuda.is_available():
    print("device name:", torch.cuda.get_device_name(0))
```

```python
model = YOLO("yolo11s.pt")

results = model.train(
    data="../data/test/ct.yaml", 
    project="../models",
    name="final_model",
    exist_ok=True,
    patience=20,
    imgsz=1024,
    single_cls=True,
    dropout=0.25,
    cos_lr=True, 
    degrees=10.0,
    scale=0.5,
    fliplr=0.5,
    mosaic=0.0,
)
```
